```python
import torch
import torch.nn as nn
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import r2_score

# Создание случайных данных для примера
np.random.seed(0)
torch.manual_seed(0)

# Генерация данных о погоде
num_samples = 1000
num_features = 5  # Количество предыдущих дней
X = np.random.randn(num_samples, num_features)  # Входные данные (температура за предыдущие дни)
y = np.random.randn(num_samples, 1)  # Выходные данные (температура на следующий день)

# Преобразование данных в тензоры PyTorch
X_tensor = torch.tensor(X, dtype=torch.float32)
y_tensor = torch.tensor(y, dtype=torch.float32)

# Определение архитектуры нейронной сети
class WeatherPredictor(nn.Module):
    def __init__(self, input_size, hidden_size, output_size):
        super(WeatherPredictor, self).__init__()
        self.fc1 = nn.Linear(input_size, hidden_size)  # Полносвязный слой
        self.fc2 = nn.Linear(hidden_size, output_size)  # Полносвязный слой

    def forward(self, x):
        x = torch.relu(self.fc1(x))  # Применение функции активации ReLU
        x = self.fc2(x)  # Выходной слой без функции активации (регрессия)
        return x

# Создание экземпляра нейронной сети
input_size = num_features
hidden_size = 10
output_size = 1
model = WeatherPredictor(input_size, hidden_size, output_size)

# Определение функции потерь и оптимизатора
criterion = nn.MSELoss()  # Функция потерь для задачи регрессии
optimizer = torch.optim.Adam(model.parameters(), lr=0.01)  # Оптимизатор

# Списки для записи потерь и коэффициента детерминации
losses = []
r2_scores = []

# Обучение модели
num_epochs = 1000
for epoch in range(num_epochs):
    # Прямой проход
    outputs = model(X_tensor)
    loss = criterion(outputs, y_tensor)
    losses.append(loss.item())  # Записываем потери
    
    # Рассчет коэффициента детерминации
    y_true = y_tensor.numpy()
    y_pred = outputs.detach().numpy()
    r2 = r2_score(y_true, y_pred)
    r2_scores.append(r2)
    
    # Обратное распространение и оптимизация
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
    
    if (epoch+1) % 100 == 0:
        print(f'Эпоха [{epoch+1}/{num_epochs}], Потери: {loss.item():.4f}, R^2: {r2:.4f}')

# Построение графика потерь
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.plot(losses, label='Потери')
plt.xlabel('Эпоха')
plt.ylabel('Потери')
plt.title('График обучения (MSE)')
plt.legend()

# Построение графика коэффициента детерминации
plt.subplot(1, 2, 2)
plt.plot(r2_scores, label='R^2')
plt.xlabel('Эпоха')
plt.ylabel('R^2')
plt.title('График обучения (R^2)')
plt.legend()

plt.tight_layout()
plt.show()

# Пример использования обученной модели для предсказания погоды
sample_input = torch.tensor(X[0], dtype=torch.float32)  # Пример входных данных (температура за предыдущие дни)
predicted_temperature = model(sample_input).item()  # Предсказанная температура на следующий день
print("Предсказанная температура на следующий день:", predicted_temperature)
```
